## Individual graphs for supply-demand factors that influence US home prices

In this Notebook i am showing the graphs for indiviual factors

#### importing all the required modules 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

#### importing all the datasets 

```python
proxy_prices = pd.read_csv('data_csv/PrHomePrice.csv')
proxy_prices = proxy_prices.rename(columns={"CSUSHPISA":"PRICE"})

gdp_percapita = pd.read_csv('data_csv/CONGDP.csv')
gdp_percapita =gdp_percapita.rename(columns={"NYGDPPCAPKDUSA":"GDPCAP"})

mortage_rate = pd.read_csv('data_csv/Mortage.csv')
mortage_rate = mortage_rate.rename(columns={"30-year fixed-rate average":"Rates"})
mortage_rate["Rates"] = mortage_rate["Rates"].replace({'%':''},regex=True)
mortage_rate.sort_values(by=["Year"],ascending=True,inplace=True)


labor_force = pd.read_csv('data_csv/Laborforce.csv')
labor_force = labor_force.rename(columns={"USCONS":"LABOR"})

permits = pd.read_csv('data_csv/PERMIT.csv')

hcai = pd.read_csv('data_csv/HCAI.csv')

inventory = pd.read_csv('data_csv/Inventory.csv')
inventory = inventory.rename(columns={"MSACSR":"INVENTORY"})

homless_calamity = pd.read_csv('data_csv/Homeless_natural_disasters.csv')
homless_calamity.sort_values(by=["YEAR"],ascending=True,inplace=True)

green_card = pd.read_csv('data_csv/Greencardiss.csv')
green_card["Number"] = green_card["Number"] / 1000
green_card.sort_values(by=["Year"],ascending=True,inplace=True)

crime_rate = pd.read_csv('data_csv/crime_rate.csv')
```

```python
#storing the datasets to access it in another notebook
%store proxy_prices
%store gdp_percapita
%store mortage_rate
%store labor_force
%store permits
%store hcai
%store inventory
%store homless_calamity
%store green_card
%store crime_rate
```

### Plotting All the data gathered separately

```python
proxy_prices.head(5)
```

```python
plt.figure(figsize=[10,6])
year = pd.date_range('2000-01-01',periods=len(proxy_prices),freq='1M')
sns.lineplot(data=proxy_prices,x=year,y='PRICE')
plt.title("House Prices per Year")
plt.xlabel("YEAR")
plt.legend(["USD Price"],loc="upper left");
```

```python
gdp_percapita.head(5)
```

```python
plt.figure(figsize=[10,6])
sns.pointplot(data=gdp_percapita,x="DATE",y='GDPCAP')
plt.xticks(rotation=60)
plt.title("GDP Per Capita per Year")
plt.legend(["USD"],loc="upper left");
```

```python
mortage_rate.head(5)
```

```python
plt.figure(figsize=[10,6])
sns.pointplot(data=mortage_rate,x="Year",y="Rates")
plt.title("Mortage Rates by Year")
plt.xlabel("YEAR")
plt.ylabel("RATE")
plt.legend(["Percentage(%)"]);
```

```python
labor_force.head(5)
```

```python
plt.figure(figsize=[10,6])
year = pd.date_range('2000-01-01',periods=len(labor_force),freq='1M')
sns.lineplot(data=labor_force,x=year,y="LABOR")
plt.title("Number Of laborer's per Year")
plt.xlabel("YEAR")
plt.legend(["Labor in Thousands"],loc="upper left");
```

```python
plt.figure(figsize=[10,6])
year = pd.date_range('2000-01-01',periods=len(permits),freq='1M')
sns.lineplot(data=permits,x=year,y="PERMIT")
plt.title("Permits given for Construction")
plt.legend(["Number of Permits"])
plt.xlabel("YEAR");
```

```python
hcai.head(5)
```

```python
plt.figure(figsize=[10,6])
year = pd.date_range('2000-01-01',periods=len(hcai),freq='Q')
sns.lineplot(data=hcai,x=year,y="Total\nRisk")
sns.lineplot(data=hcai,x=year,y="Borrower\nRisk")
sns.lineplot(data=hcai,x=year,y="Product\nRisk")
legend = plt.legend(["Total Risk","Borrower's Risk","Product Risk"])
legend.set_title("Risk")
plt.title("Home Credit Availibility Index")
plt.ylabel("RISK")
plt.xlabel("YEAR");
```

```python
plt.figure(figsize=[10,6])
year = pd.date_range('2000-01-01',periods=len(inventory),freq='1M')
sns.lineplot(data=inventory,x=year,y="INVENTORY")
plt.title("Number of Houses Available For Sale")
plt.xlabel("YEAR");
```

```python
homless_calamity.head(5)
```

```python
plt.figure(figsize=[10,6])
sns.pointplot(data=homless_calamity,x="YEAR",y="HOMELESS")
plt.title("Number of People geting Homeless by Calamities per Year")
plt.legend(["Number of People"],loc="upper left")
```

```python
plt.figure(figsize=[10,6])
sns.pointplot(data=green_card,x="Year",y="Number")
plt.xlabel("YEAR")
plt.ylabel("NUMBERS")
plt.xticks(rotation=15)
plt.title("Number of Green Cards Issued per Year")
plt.legend(["Issued in Thousands"]);
```
