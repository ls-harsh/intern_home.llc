## Data science model for supply-demand factors that influence US home prices

In this notebook i am showing the relation between the factors and how its affecting the prices

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Accessing the datasets from main notebook
%store -r proxy_prices
%store -r gdp_percapita
%store -r mortage_rate
%store -r labor_force
%store -r permits
%store -r hcai
%store -r inventory
%store -r homless_calamity
%store -r green_card
%store -r crime_rate
%store -r sold
```

# Supply factors affecting house prices 

#### Graphs showing the relation b/w Number of houses sold and average prices of houses by respective year  

```python
plt.figure(figsize=[15,6])
plt.subplot(1,2,1)
year = pd.date_range('2000',periods=len(sold),freq='1Y')
sns.lineplot(data=sold,x=year,y="Total sold")
plt.xlabel("YEAR")
plt.title("Number of Houses sold in Millions")
plt.legend(["In Millions"],loc="upper left")
plt.subplot(1,2,2)
year = pd.date_range('2000-01-01',periods=len(proxy_prices),freq='1M')
sns.lineplot(data=proxy_prices,x=year,y='PRICE')
plt.xlabel("YEAR")
plt.title("Average house Prices per Year")
plt.legend(["USD"],loc="upper left");
```

#### Graphs showing the relation b/w Permits given and average prices of  houses by respective year 

```python
plt.figure(figsize=[15,6])
plt.subplot(1,2,1)
year = pd.date_range('2000-01-01',periods=len(permits),freq='1M')
sns.lineplot(data=permits,x=year,y="PERMIT")
plt.xlabel("YEAR")
plt.title("Permits given for Construction")
plt.legend(["In Thousands"],loc="upper left")
plt.subplot(1,2,2)
year = pd.date_range('2000-01-01',periods=len(proxy_prices),freq='1M')
sns.lineplot(data=proxy_prices,x=year,y='PRICE')
plt.xlabel("YEAR")
plt.title("Average house Prices per Year")
plt.legend(["USD"],loc="upper left");
```

# Demand Factors affecting the house prices

```python
plt.figure(figsize=[15,6])
plt.subplot(1,2,1)
year = pd.date_range('2000-01-01',periods=len(gdp_percapita),freq='1Y')
sns.lineplot(data=gdp_percapita,x=year,y="GDPCAP")
plt.xlabel("YEAR")
plt.ylabel("GDP per Capita")
plt.title("GDP Per Capita per Year")
plt.legend(["USD"],loc="upper left")
plt.subplot(1,2,2)
year = pd.date_range('2000-01-01',periods=len(proxy_prices),freq='1M')
sns.lineplot(data=proxy_prices,x=year,y='PRICE')
plt.xlabel("YEAR")
plt.title("Average house Prices per Year")
plt.legend(["USD"],loc="upper left");
```

#### Graph showing the relation between Mortage rate and Average prices of house by respective year

```python
plt.figure(figsize=[15,6])
plt.subplot(1,2,1)
year = pd.date_range('2000',periods=len(mortage_rate),freq='1Y')
sns.lineplot(data=mortage_rate,x=year,y="Rates")
plt.title("Mortage Rates by Year")
plt.xlabel("YEAR")
plt.ylabel("30-year fixed-rate")
plt.legend(["Percentage(%)"])
plt.subplot(1,2,2)
year = pd.date_range('2000-01-01',periods=len(proxy_prices),freq='1M')
sns.lineplot(data=proxy_prices,x=year,y='PRICE')
plt.xlabel("YEAR")
plt.title("Average house Prices per Year")
plt.legend(["USD"],loc="upper left");
```

#### Graph showing the relation between Home Credit Availibility Index and Average prices of house by respective year

```python
plt.figure(figsize=[15,6])
plt.subplot(1,2,1)
year = pd.date_range('2000-01-01',periods=len(hcai),freq='Q')
sns.lineplot(data=hcai,x=year,y="Total\nRisk")
sns.lineplot(data=hcai,x=year,y="Borrower\nRisk")
sns.lineplot(data=hcai,x=year,y="Product\nRisk")
legend = plt.legend(["Total Risk","Borrower's Risk","Product Risk"])
legend.set_title("Risk")
plt.title("Home Credit Availibility Index")
plt.ylabel("RISK")
plt.xlabel("YEAR");
plt.subplot(1,2,2)
year = pd.date_range('2000-01-01',periods=len(proxy_prices),freq='1M')
sns.lineplot(data=proxy_prices,x=year,y='PRICE')
plt.xlabel("YEAR")
plt.title("Average house Prices per Year")
plt.legend(["USD"],loc="upper left");
```

# Conclusion from the above graphs  

We can see more trends on downside during 2008-2010 years due to Economic Crisis which affected the housing industry a lot and we can an econmic Recession during that period.

We can also see that in supply Factors steep Downfall during the same time period whether it is Number of houses sold or permits given or even drop in the work force for the construction of the houses.


From the Demand graphs we can notice that there's a high rise in the number of houses available for sales,but at same time we can see downfall in HCAI,GDP per Capita of the nation but the intresting part is that there is changes in the Mortage Rates and it remains Constant.

Another trend can be seen in majority of graphs in year 2020 and onwards and it's due to the Covid-19 or even due to the NationWide Lockdown and Restrictions put on people. 

Other factors affecting the house prices are natural calmities,green card issue/permanant citizenship being granted or even in the crime rates.

There are even many other minor factors not included like size of house,social mindset of people and locality,and other factors as they are difficult to be recoreded or dont affect much.
